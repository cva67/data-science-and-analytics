```python
### The Framework
#1. State **H₀** (null hypothesis) and **H₁** (alternative)
#2. Choose significance level **α** (typically 0.05)
#3. Select the right test for your data type
#4. Compute the **p-value**
#5. Reject H₀ if p-value < α
```

```python
#“Does age affect survival?”
#H₀: Age has no effect on survival
#H₁: Age affects survival
survived = df[df['Survived'] == 1]['Age']
not_survived = df[df['Survived'] == 0]['Age']

# Perform independent t-test
t_stat, p_value = ttest_ind(survived, not_survived)

print("T-statistic:", t_stat)
print("P-value:", p_value)

# Interpretation
if p_value < 0.05:
    print("Reject H0: Age affects survival")
else:
    print("Fail to reject H0: No evidence age affects survival")
```

```python
#“Did gender affect survival?”
#Hypotheses Null Hypothesis 
#H₀: Gender has no effect on survival. 
#Alternative Hypothesis 
#H₁: Gender affects survival.

contingency = pd.crosstab(df['Sex'], df['Survived'])
contingency.columns = ['Did Not Survive', 'Survived']
print('Contingency Table:')
print(contingency)

chi2, p_chi, dof, expected =chi2_contingency(contingency)

print(f'\n── Chi-Square Test: Sex vs Survival ──')
print(f'  Chi² statistic : {chi2:.4f}')
print(f'  Degrees of freedom: {dof}')
print(f'  p-value        : {p_chi:.2e}')

if p_chi < 0.05:
    print('\n  ✅ Reject H₀. Sex and survival ARE significantly associated.')
else:
    print('\n  ❌ Fail to reject H₀.')

```
