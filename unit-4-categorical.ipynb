```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder, OneHotEncoder
```

```python
df = pd.read_csv('PlayTennis.csv')
df.head()
```

```python
df.isnull().sum()* 100 / len(df)
```

```python
df['Temperature'].unique()
```

```python
ordinal_encoder = OrdinalEncoder(
    categories=[['Cool', 'Mild', 'Hot']]
)
```

```python
df['Temperature'] = ordinal_encoder.fit_transform(
    df[['Temperature']]
)
```

```python
df.head()
```

```python
df['Wind'].unique()
```

```python
le_wind = LabelEncoder()
df['Wind'] = le_wind.fit_transform(df['Wind'])
```

```python
print(df[['Wind']].head())
```

```python
df.head()
```

```python
encoder = OneHotEncoder(sparse_output=False)

encoded = encoder.fit_transform(df[['Outlook']])

encoded_df = pd.DataFrame(
    encoded,
    columns=encoder.get_feature_names_out(['Outlook'])
)
df = df.drop('Outlook', axis=1).join(encoded_df)
```

```python
df.head()
```

```python
for col in [ 'Wind', 'Humidity']:
    freq = df[col].value_counts(normalize=True)
    df[col + '_Freq'] = df[col].map(freq)
```

```python
le_target = LabelEncoder()
df['Play Tennis'] = le_target.fit_transform(df['Play Tennis'])
```
